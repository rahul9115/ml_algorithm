```python
def cluster_points(X,mu):
    clusters={}
    for x in X:
        bestmukey=min([(i[0],np.linalg.norm(x-mu[i[0]])) for i in enumerate(mu)],key=lambda x:x[1])[0]
        print("this",min([(i[0],np.linalg.norm(x-mu[i[0]])) for i in enumerate(mu)],key=lambda x:x[1]))
        print("mu",mu)
        
        try:
            clusters[bestmukey].append(x)
        except:
            clusters[bestmukey]=[x]
    return clusters
    
def reevaluate_centers(mu,clusters):
    newmu=[]
    keys=sorted(clusters.keys())
    for i in keys:
        newmu.append(np.mean(clusters[i],axis=0))
    return newmu

    
def has_converged(old_mu,mu):
    return set([tuple(i) for i in old_mu])==set([tuple(i) for i in mu])

def find_centers(X,k):
    old_mu=random.sample(X,k)
    print(old_mu)
    
    mu=random.sample(X,k)
    clusters={}
    while not has_converged(old_mu,mu):
        old_mu=mu
        clusters=cluster_points(X,mu)
        print(type(clusters))
        mu=reevaluate_centers(old_mu,clusters)
    return clusters

```

```python
clusters=find_centers(data,3)
```

```python
output=[]

for i,j in clusters.items():
    for k1 in j:
        k1=k1.tolist()
        k1.append(i)
        k1=np.asanyarray(k1)
        
        output.append(k1)
    


df1=pd.DataFrame(data=output,columns=['experience_level','job_title','company_size','salary_in_usd','clusters'])
```
